```python
def my_kmeans(image,k):
    #k = number of clusters
    
    # Reshaping the image  
    pixel_vals = image.reshape((-1,1)) 

    # Convert to float type only for supporting cv2.kmean
    pixel_vals = np.float32(pixel_vals)
    criteria   = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85) #criteria
    retval, labels, centers = cv2.kmeans(pixel_vals, k, 
                None, criteria, 10, cv2.KMEANS_PP_CENTERS) 
    centers = np.uint8(centers) # convert data into 8-bit values 
    
    segmented_data   = centers[labels.flatten()] # Mapping labels to center points( RGB Value)
    segmented_image  = segmented_data.reshape((image.shape)) # reshape data into the original image dimensions

    segmented_labels = labels # Mapping labels to center points( RGB Value)
    segmented_labels = segmented_labels.reshape((image.shape)) # reshape data into the original image dimensions
    
    return segmented_image,segmented_labels
```

# K_means implementation

## Pre-processing

First, we perform a preprocessing :
- A normalization
- A threshold composed of a local threshold, and a computed one with the average in the center of the image  

This allows to exclude many regions so that the image can be better processed by k_means

```python
def normalisation(image) :
    min_val = np.min(image)
    max_val = np.max(image)
    return (image - min_val)/(max_val - min_val)
```

```python
def thresh(image):
    
    filtered_image = ndi.gaussian_filter(image, sigma=(1, 1), order=0)
    
    n,m = image.shape
    binary1 = filtered_image > threshold_local(image, 9)
    
    thresh2 = np.mean(image[(2*n//4):(3*n//4),(2*m//4):(3*m//4)])*9/10
    
    binary2          = morphology.remove_small_holes(
                        morphology.remove_small_objects(
                          filtered_image > thresh2, 
                            min_size=500,connectivity=1),
                             area_threshold=500)
    binary2 = morphology.opening(binary2, morphology.disk(3))
    binary2 = binary2.astype('uint8')
    
    binary_image = (binary1 + binary2 == 2)
    binary_image = binary_image.astype('uint8')
    return binary_image
```

## k_means algorithm

```python
def k_means(image, nber_clusters = 10) :
    
    image_n = normalisation(image)
    Seg_thresh = thresh(image_n)

    Seg_km,Seg_km_labels = my_kmeans(image_n,nber_clusters)

    # Get mean pixel intensity values under all Kmeans labels
    Img_label_means = np.empty(nber_clusters)
    for i in range(0,nber_clusters):
        Img_label_mask = Seg_km_labels==i
        Img_label_mask = Img_label_mask.astype('uint8')
        tmp            = cv2.mean(Img_test, Img_label_mask)
        Img_label_means[i] = tmp[0]
    #print(Img_label_means)

    # Set a target intensity value and get the Kmeans label closest to it
    #Target_value = np.mean(image[Seg_thresh == 1])
    Target_value = cv2.mean(Img_test, Seg_thresh)
    Target_value = Target_value[0]
    Diff = np.absolute(Img_label_means-Target_value*np.ones(nber_clusters))
    Label_select = np.argmin(Diff)

    # Filter the selected label and get a binary segmentation mask
    Img_label_select  = Seg_km_labels==Label_select
    Img_label_select  = Img_label_select.astype('float64')

    # Examples to refine your segmentation mask
    Img_label_select_smooth = gaussian(Img_label_select, 2, 
                                preserve_range=True)

    Img_label_select_smooth_clean = morphology.remove_small_holes(
        morphology.remove_small_objects(
            Img_label_select_smooth>0.45, min_size=50,connectivity=1),
        area_threshold=50)
    
    return Img_label_select_smooth_clean
```

## Post_processing. 

For the k_means, we choose to keep the related component which, among the 5 largest related components, is the highest in the image. This allows us to easily localize the corpus callosum.

```python
def keep_top_components_kmeans(binary_image):
    # Step 1: Label connected components in the binary image
    labeled_image = label(binary_image)
    regions = regionprops(labeled_image)
    
    # Step 2: Sort regions by area
    regions.sort(key=lambda x: x.area, reverse=True)
    
    # Step 3: Keep only the 5 largest regions
    regions = regions[:5]
    
    # Step 4: Sort regions by height
    regions.sort(key=lambda x: x.bbox[3] - x.bbox[1], reverse=True)
    
    # Step 5: Keep only the region with the highest height
    region_to_keep = regions[0]
    
    # Step 6: Create a new binary image with only the kept region
    output_image = np.zeros_like(binary_image)
    for coord in region_to_keep.coords:
        output_image[coord[0], coord[1]] = 1
        
    return output_image
```

## Quality metrics 
I implemented jaccard index and dice coefficient, which are quality metrics that are able to compare the segmentation provided by the implementation to the provided ground truth

```python
def dice_coefficient(binary_image1, binary_image2):
    intersection = np.logical_and(binary_image1, binary_image2)
    return 2 * intersection.sum() / (binary_image1.sum() + binary_image2.sum())
```

```python
def jaccard_index(binary_image1, binary_image2):
    # Compute the intersection of the two binary images
    intersection = np.logical_and(binary_image1, binary_image2)
    # Compute the union of the two binary images
    union = np.logical_or(binary_image1, binary_image2)
    # Compute the Jaccard index
    jaccard_index = np.sum(intersection) / np.sum(union)
    return jaccard_index
```
